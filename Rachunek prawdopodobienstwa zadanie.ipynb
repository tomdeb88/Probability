```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
x = np.linspace(0, 6, 100)
y = np.linspace(0, 6, 10000)
```

```python
q0 = np.quantile(x, 0.0)
q1 = np.round(np.quantile(x, 0.25))
q3 = np.round(np.quantile(x, 0.75))
q4 = np.round(np.quantile(x, 1.0))
```

```python
def gauss(x, sigma=1, mu=0):
    return stats.norm.pdf(x, mu, sigma)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x, gauss(x, sigma=1,mu=3), label='odchylenie standardowe = 1, rednia = 3,set=100')
plt.plot(y, gauss(y, sigma=1,mu=3),label='odchylenie standardowe = 1, rednia = 3,set=10000',ls=':',lw=5,c='yellow',alpha=0.5)
plt.axvline(x=3,c='r',label='Srednia, Mediana, Moda')
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper left')
plt.show()
```

### Oba zbiory maja taka sama srednia,mediane i mode poniewaz sa to rozklady symetryczne
